## Definição do número de agrupamentos usando o método Elbow (método do cotovelo) 

- Antes da aplicação do algoritmo k-means, precisamos informar o número de clusters (grupos) que deverão ser formados 
- Quando não há uma definição prévia por parte da empresa da quantidade de grupos que deverão ser formados, para encontrarmos o número ideal de clusters, podemos usar o método chamado elbow (cotovelo).
- Esse método roda o KMeans para quantidades diferentes de clusters e diz qual dessas quantidades é o número ótimo de clusters. 
- O que geralmente acontece ao aumentar a quantidade de clusters no KMeans é que as diferenças entre clusters se tornam muito pequenas, e as diferenças das observações intra-clusters vão aumentando. 
- Assim, é preciso achar um equilíbrio em que as observações que formam cada agrupamento sejam o mais homogêneas possível e que os agrupamentos formados sejam o mais diferentes um dos outros.


```
# Iremos criar uma função que retorne uma lista com o valor do wcss/inertia (média das distâncias de cada ponto em relação ao centro do cluster)
# Para cada quantidade de clusters (1 a 19) 

def calcular_wcss(dataset):
  wcss = []
  range_values = range(1, 20)
  for i in range_values:
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(creditcard_df)
    wcss.append(kmeans.inertia_)
  return wcss
```

```
lista_wcss = calcular_wcss(creditcard_df)
```

```
# Usaremos um gráfico de linha para poder visualizar a relação entre o valor do wcss/inertia para cada quantidade de clusters

plt.plot(lista_wcss, 'bx-')
plt.title('Método cotovelo para o cálculo do número ideal de clusters')
plt.xlabel('Clusters')
plt.ylabel('WCSS');
```
